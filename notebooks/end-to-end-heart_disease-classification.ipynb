```python
#data dictionary
# 1. age – Age of the patient (years)
# 2. sex – Gender (0 = Female, 1 = Male)
# 3. cp – Chest pain type (1 = Typical angina, 2 = Atypical angina, 3 = Non-anginal pain, 4 = Asymptomatic)
# 4. trestbps – Resting blood pressure (mm Hg)
# 5. chol – Serum cholesterol (mg/dl)
# 6. fbs – Fasting blood sugar > 120 mg/dl (0 = False, 1 = True)
# 7. restecg – Resting electrocardiographic results (0 = Normal, 1 = ST-T abnormality, 2 = Left ventricular hypertrophy)
# 8. thalach – Maximum heart rate achieved
# 9. exang – Exercise induced angina (0 = No, 1 = Yes)
# 10. oldpeak – ST depression induced by exercise relative to rest
# 11. slope – Slope of peak exercise ST segment (1 = Upsloping, 2 = Flat, 3 = Downsloping)
# 12. ca – Number of major vessels colored by fluoroscopy (0–3)
# 13. thal – Thalassemia test result (3 = Normal, 6 = Fixed defect, 7 = Reversible defect)
# 14. target – Diagnosis of heart disease (0 = No disease, 1 = Disease)
```

```python
#Split the data
X=df.drop("target",axis=1)
y=df["target"]
X,y
```

```python
np.random.seed(42)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train
```

```python
models={"logisticreg":LogisticRegression(),"KNN":KNeighborsClassifier(),"randomForest":RandomForestClassifier()}
def fit_and_score(models,X_train,X_test,y_train,y_test):
    np.random.seed(42)
    model_scores={}
    for name,model in models.items():
        model.fit(X_train,y_train)
        model_scores[name]=model.score(X_test,y_test)
    return model_scores
        


```

```python
model_scores=fit_and_score(models=models,X_train=X_train,X_test=X_test,y_train=y_train,y_test=y_test)
model_scores
```

```python
model_compare=pd.DataFrame(model_scores,index=["accuracy"])
model_compare.T.plot.bar();
```

```python
#tuning KNN
train_scores=[]
test_scores=[]
neighbors=range(1,21)
knn=KNeighborsClassifier()
for i in neighbors:
    knn.set_params(n_neighbors=i)
    knn.fit(X_train,y_train)
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
    
```

```python
train_scores
```

```python
test_scores
```

```python
plt.plot(neighbors,train_scores,label="Train score")
plt.plot(neighbors,test_scores,label="Test score")
plt.xticks(np.arange(1,21,1))
plt.xlabel("neighbors")
plt.ylabel("Model score")
plt.legend()
print(f"Maximum KNN scores on the test data: {max(test_scores)*100:.2f}%")
```

```python
###Hyperparameter tuning with Randomized searchCV
#hyperparmater grid for logistic regression grid
logreg_grid={"C":np.logspace(-4,4,20),
            "solver":["liblinear"]}
#hyperparameter grid for Random forest classfier
rf_grid = {
    "n_estimators": np.arange(10, 1000, 50),
    "max_depth": [None, 3, 5, 10],
    "min_samples_split": np.arange(2, 20, 2),
    "min_samples_leaf": np.arange(1, 20, 2)
}
```

```python
#tuning logistic regression
np.random.seed(42)
rs_logreg=RandomizedSearchCV(LogisticRegression(),param_distributions=logreg_grid,
                             cv=5,n_iter=20,verbose=True)
rs_logreg.fit(X_train,y_train)
```

```python
rs_logreg.best_params_
```

```python
rs_logreg.score(X_test,y_test)
```

```python
###as wehave tuned logistic regresion now its time for Randomforstclassfier
```

```python
rs_rf = RandomizedSearchCV(
    RandomForestClassifier(),
    param_distributions=rf_grid,
    cv=5,
    n_iter=20,
    verbose=True
)
rs_rf.fit(X_train, y_train)
```

```python
#find the best hyperparameters
rs_rf.best_params_
```

```python
rs_rf.score(X_test,y_test)
```

```python
##Hyperparameter tuning with grid search cv
log_reg_grid={"C":np.logspace(-4,4,30),
             "solver":["liblinear"]}
#setting gridhyperparameter for log reg
gs_log_reg=GridSearchCV(LogisticRegression(),param_grid=log_reg_grid,
                       cv=5,verbose=True)
#fit
gs_log_reg.fit(X_train,y_train)
```

```python
gs_log_reg.score(X_test,y_test)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# The correct parameter to scale font elements is 'font_scale'
sns.set(font_scale=1.5)

def plot_conf_mat(y_test, y_preds):
    """
    Plots an improved and visually appealing confusion matrix.
    """
    fig, ax = plt.subplots(figsize=(4, 4))
    cm = confusion_matrix(y_test, y_preds)
    
    # Use the 'ax' parameter to tell seaborn where to draw the heatmap
    sns.heatmap(cm,
                ax=ax,
                annot=True,      # Annotate cells with their values
                fmt='d',         # Format annotations as integers
                cbar=False,      # Remove the color bar
                cmap='Blues')    # Use a visually clear color map
    
    # Set labels and a title using the axes object for better practice
    ax.set_xlabel("Predicted Label")
    ax.set_ylabel("True Label")
    ax.set_title("Confusion Matrix")
    
    plt.show() # Explicitly display the plot

# Call the function with the sample data
plot_conf_mat(y_test, y_preds)
```
